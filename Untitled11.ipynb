```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('C:\\Users\\Kunal Dhariwal\\Downloads\\city_weather.csv')
```

```python
data.drop(['Unnamed: 0','date_txt','lat','lon','country','temp_max','temp'], axis=1, inplace=True)
```

```python
data.head()
```

```python
y = data.temp_min
X = data.drop(['name', 'clouds','rain', 'wind_direction_deg', 'wind_speed', 'temp_min','pressure'], axis=1)
```

```python
X.head()
```

```python
data.columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2)
print("\nX_train:\n")
print(X_train.head())
print(X_train.shape)

print("\nX_test:\n")
print(X_test.head())
print(X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
reg.score(X_test,y_test)
```

```python
y_pred = reg.predict(X_test)
```

```python
y_pred
```

```python
from sklearn import ensemble
clf = ensemble.GradientBoostingRegressor(n_estimators = 400, max_depth = 5, min_samples_split = 2,
          learning_rate = 0.1, loss = 'ls')
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test,y_test)
```
